```python
#1. Escribe una función llamada convertir_celsius_a_fahrenheit que reciba un valor en grados Celsius y devuelva su equivalente en Fahrenheit.

def convertir_celsius_a_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Solicitar al usuario que ingrese el valor en Celsius
celsius = float(input("Ingresa el valor en grados Celsius: "))
fahrenheit = convertir_celsius_a_fahrenheit(celsius)
print(f"{celsius} grados Celsius son {fahrenheit} grados Fahrenheit.")
```

```python
#2. Crea una función calcular_promedio que reciba una lista de números y devuelva el promedio. Valida que la lista no esté vacía.

def calcular_promedio(numeros):
    if not numeros:
        return "La lista está vacía, no se puede calcular el promedio."
    return sum(numeros) / len(numeros)

# Solicitar al usuario que ingrese 10 números
numeros = []
print("Ingresa 10 números:")
for _ in range(10):
    numero = float(input("Ingresa un número: "))
    numeros.append(numero)

# Mostrar los números ingresados
print(f"Números ingresados: {numeros}")

promedio = calcular_promedio(numeros)
print(f"El promedio de la lista es: {promedio}")
```

```python
#3. Desarrolla una función llamada calcular_factorial que calcule el factorial de un número entero positivo utilizando recursión. Asegúrate de manejar casos inválidos.

def calcular_factorial(n):
    # Manejar casos inválidos
    if not isinstance(n, int) or n < 0:
        return "Error: El número debe ser un entero positivo."

    # Caso base
    if n == 0 or n == 1:
        return 1

    # Llamada recursiva
    return n * calcular_factorial(n - 1)

# Solicitar al usuario que ingrese un número
try:
    numero = int(input("Ingresa un número entero positivo: "))
    resultado = calcular_factorial(numero)
    print(f"El factorial de {numero} es {resultado}")
except ValueError:
    print("Error: Debes ingresar un número entero.")
```

```python
#5. Crea una función llamada generar_factura que reciba una lista de precios y un porcentaje de descuento opcional. Calcula el total después de aplicar el descuento.

def generar_factura():
    precios = []
    while True:
        precio = input("Ingresa un precio (o 'fin' para terminar): ")
        if precio.lower() == 'fin':
            break
        try:
            precios.append(float(precio))
        except ValueError:
            print("Por favor, ingresa un número válido.")
    
    descuento = input("Ingresa el porcentaje de descuento: ")
    try:
        descuento = float(descuento)
    except ValueError:
        print("Por favor, ingresa un número válido para el descuento.")
        return
    
    total = sum(precios)
    total_con_descuento = total * (1 - descuento / 100)
    
    print(f"El total antes de aplicar el descuento es: {total:.2f}")
    print(f"Se aplicó un descuento del {descuento}%")
    print(f"El total después de aplicar el descuento es: {total_con_descuento:.2f}")

# Llamada a la función
generar_factura()
```

```python
#6. Implementa una función llamada configurar_envio que reciba múltiples argumentos mediante **kwargs. Esta función debe permitir configurar parámetros como dirección, método de envío y tiempo estimado.

def configurar_envio(**kwargs):
    direccion = kwargs.get('direccion', 'No especificada')
    metodo_envio = kwargs.get('metodo_envio', 'Estándar')
    tiempo_estimado = kwargs.get('tiempo_estimado', 'Desconocido')

    configuracion = {
        'Direccion': direccion,
        'Metodo de Envio': metodo_envio,
        'Tiempo Estimado': tiempo_estimado
    }

    return configuracion

# Ejemplo de uso
parametros_envio = configurar_envio(direccion=input("Ingrese la dirección: "),metodo_envio= input("Ingrese el método de envío (Estándar/Express): "), tiempo_estimado=input("Ingrese el tiempo estimado de entrega: "))
print(parametros_envio)
```

```python
#7. Crea un programa que lea un archivo CSV llamado productos.csv e imprima los nombres de los productos y sus precios.

import csv

# Abrir el archivo CSV
with open('productos.csv', mode='r') as file:
    # Crear un objeto lector de CSV
    csv_reader = csv.reader(file)
    
    # Saltar la fila de encabezado
    next(csv_reader)
    
    # Iterar sobre cada fila en el archivo CSV
    for row in csv_reader:
        # Imprimir el nombre del producto y el precio
        print(f"Producto: {row[0]}, Precio: {row[1]}")
```

```python
#8. Crea una aplicación que lea un archivo CSV con información de empleados (nombre, edad, salario) y genere un reporte binario que contenga solo los empleados con salario superior a un umbral dado.

import csv
import pickle

def generate_report(input_csv, output_bin, salary_threshold):
    employees = []

    # Leer el archivo CSV
    with open(input_csv, mode='r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Saltar la fila de encabezado

        # Filtrar empleados según el umbral de salario
        for row in csv_reader:
            name, age, salary = row
            if float(salary) > salary_threshold:
                employees.append({'name': name, 'age': int(age), 'salary': float(salary)})

    # Escribir los empleados filtrados en un archivo binario
    with open(output_bin, mode='wb') as file:
        pickle.dump(employees, file)

# Ejemplo de uso
generate_report('empleados.csv', 'reporte.bin', 50000)
```
